```python
import pandas as pd
import numpy as np
data = pd.read_csv('diabetes.csv')
data.head(5)
```

```python
data.iloc[:,0:8] = data.iloc[:,0:8].replace(0,np.nan)
data.head(5)
```

```python
data.iloc[:,0:4] = data.iloc[:,0:4].fillna(method = 'bfill')
data.iloc[:,5:8] = data.iloc[:,5:8].fillna(method = 'bfill')
```

```python
data = data.dropna()
data.head(5)
```

```python
X = data.iloc[:,0:8].values
y = data.iloc[:,8].values
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()
model.add(Dense(units = 9, input_dim = 8, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units = 5,  kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(units= 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
model.fit(X,y, epochs = 100, batch_size = 10, verbose = 1)
```

```python
scores = model.evaluate(X, y)
print("Accuracy: %.2f" % (scores[1]*100))
```
